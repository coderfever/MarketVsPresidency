```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt

with urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_Presidents_of_the_United_States')as url:
    s = url.read()
soup = BeautifulSoup(s, 'html.parser')
table = soup.find('table', class_='wikitable')
rows = table.findAll('tr')
rows = rows[2:]
```

```python
# cleanup presidents table
data = []
for row in rows:
    tds = row.findAll('td')
    if (len(tds) > 6):
        text = [td.text for td in tds[0:7]]
        data.append(text)
columns = ['index', 'term', 'a', 'President', 'Prior Office', 'b', 'Party']
df = pd.DataFrame(data, columns = columns)
df.drop(['a','b'], 1, inplace=True)
df = df.set_index('index')
df['term'] = df['term'].str.replace('\[[a-z]\]', repl='')
df['term'] = df['term'].str.replace('\\n\\n', repl='\\n')
terms = df['term'].str.split('\n', expand=True)
df.drop('term', axis=1, inplace=True)
mask = terms.iloc[:,0] == ''
terms.loc[mask, :] = terms.loc[mask, : ].shift(-1, axis=1)
terms.drop([1,3,4], axis=1, inplace=True)
```

```python
terms.columns = ['Start', 'End']
df = df.join(terms)
df.head()
```

```python
df['President'] = df['President'].str.split('\n', expand=True)[0]
df = df[['Start', 'End', 'President', 'Party']].copy()
df['End'][-1] = 'December 29, 2017'
df[['Start', 'End']] = df[['Start', 'End']].apply(pd.to_datetime, format='%B %d, %Y')
filtered = df[df['Start'] > '1940-01-01'].copy()
```

```python
filtered.reset_index(inplace=True)
filtered.drop('index', axis=1, inplace=True)
filtered.head()
```

```python
stocks = pd.read_csv('^GSPC.csv')
stocks = stocks[['Date', 'Close']]
stocks['Date'] = stocks['Date'].apply(pd.to_datetime, format='%Y-%m-%d')
```

```python
def which_president(stocksdf, presidentsdf):
    for index, row in presidentsdf.iterrows():
        mask = (stocksdf['Date'] > row['Start']) & (stocksdf['Date'] < row['End'])
        stocksdf.loc[mask, 'President'] = row['President']
            
    return stocksdf

newdf = which_president(stocks, df)
newdf.head()
idmin = newdf.groupby('President')['Date'].idxmin()
idmax = newdf.groupby('President')['Date'].idxmax()
```

```python
final = newdf.loc[idmin]
final = pd.concat([final, newdf.loc[idmax]])
presidents = pd.unique(final.sort_values('Date')['President'])
presidents
end = newdf.loc[idmax]['Close'].sort_index().reset_index()
start = newdf.loc[idmin]['Close'].sort_index().reset_index()
end.drop('index', axis=1, inplace=True); start.drop('index', axis=1, inplace=True); 
combined = start.join(end, rsuffix='_end')
result = filtered.join(combined)
result
```

```python
def calc_pct_change(row):
    diff = (row['Close_end'] - row['Close'])
    return round(diff / row['Close'] * 100, 1)
result['% Change'] = result.apply(calc_pct_change, axis=1)
```

```python
result = result.set_index('President')
result
```

```python
def barcolors(val):
    kc = {'Democratic': 'b', 'Republican': 'r'}
    return kc[val]
colors = list(map(barcolors, result['Party']))

# fig, axes = plt.subplots(nrows=2, ncols=1)
ax = result.plot(y='% Change', kind='bar', color=colors, legend=None)
stocks.plot(x='Date', y='Close' )
ax.set_ylabel('% Change')
plt.show()
```
